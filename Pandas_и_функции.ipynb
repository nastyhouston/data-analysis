Задание 1. Напишите функцию, которая классифицирует фильмы из материалов занятия по следующим правилам:
- оценка 2 и меньше - низкий рейтинг
- оценка 4 и меньше - средний рейтинг
- оценка 4.5 и 5 - высокий рейтинг

Результат классификации запишите в столбец class

```python
ratings = pd.read_csv('ratings.csv')
ratings.head()
```

```python
def classification(row):
  """Классифицирует фильмы в зависимости от оценки пользователей"""
  if row['rating'] <= 2.0:
    return 'низкий рейтинг'
  if row['rating'] > 2.0 and row['rating'] <= 4.0:
    return 'средний рейтинг'
  if row['rating'] > 4.0:
    return 'высокий рейтинг'
```

```python
ratings['class'] = ratings.apply(classification, axis=1)
```

```python
ratings.head()
```

Задание 2. Необходимо написать гео-классификатор, который каждой строке сможет выставить географическую принадлежность определенному региону. Т. е. если поисковый запрос содержит название города региона, то в столбце ‘region’ пишется название этого региона. Если поисковый запрос не содержит названия города, то ставим ‘undefined’.
Результат классификации запишите в отдельный столбец region.

```python
geo_data = {'Центр': ['москва', 'тула', 'ярославль'],
            'Северо-Запад': ['петербург', 'псков', 'мурманск'],
            'Дальний Восток': ['владивосток', 'сахалин', 'хабаровск']
            }
```

```python
keywords = pd.read_csv('keywords.csv')
```

```python
def geo_classification(row):
  """Определяет географический регион поискового запроса"""
  for city in sum(geo_data.values(), []):
    if city in row['keyword']:
      for region, cities in geo_data.items():
        if city in cities:
          return region
  return 'undefined'
```

```python
keywords['region'] = keywords.apply(geo_classification, axis=1)
```

```python
keywords.head()
```

```python
keywords.value_counts('region')
```

Задание 3. Проверить верно ли, что с ростом года выпуска фильма его средний рейтинг становится ниже.

```python
ratings = pd.read_csv('ratings.csv')
movies = pd.read_csv('movies.csv')
```

1. В переменную years запишите список из всех годов с 1950 по 2010.

```python
years = range(1950, 2011)
```

2. Напишите функцию production_year, которая каждой строке из названия фильма выставляет год выпуска. Не все названия фильмов содержат год выпуска в одинаковом формате, поэтому используйте следующий алгоритм:

> для каждой строки пройдите по всем годам списка years

> если номер года присутствует в названии фильма, то функция возвращает этот год как год выпуска

> если ни один из номеров года списка years не встретился в названии фильма, то возвращается 1900 год

```python
def production_year(row):
  """Определяет год выпуска фильма"""
  for year in years:
    if str(year) in row['title']:
      return year
  return 1900
```

3. Запишите год выпуска фильма по алгоритму пункта 2 в новый столбец ‘year’

```python
movies['year'] = movies.apply(production_year, axis=1)
movies.head()
```

```python
movies.value_counts('year')
```

4. Посчитайте средний рейтинг всех фильмов для каждого значения столбца ‘year’ и отсортируйте результат по убыванию рейтинга

```python
movies_ratings = ratings.merge(movies, left_on='movieId', right_on='movieId', how='inner')
movies_ratings.head()
```

```python
(
    movies_ratings.groupby('year')
    .agg({'rating': 'mean'})
    .sort_values(by='rating', ascending=False)
    )
```
